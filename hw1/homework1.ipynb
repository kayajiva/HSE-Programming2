```python
import random
import re

#эта функция предлагает пользователю выбрать тему и открывает для неё список слов
#а также обходит ошибки ввода
def choose_topic():
     topic = input('Введите название одной из тем: "цветы", "животные" или "цвета": ').lower()
     while topic != 'животные' and topic != 'цветы' and topic != 'цвета':
         topic = input('Пожалуйста, выберите только одну из трех тем: ').lower()
     else:
      if topic == 'животные':
         fname = 'animals.txt'
      elif topic == 'цветы':
         fname = 'flowers.txt'
      elif topic == 'цвета':
         fname = 'colors.txt'
     with open(fname, 'r', encoding = 'utf-8') as f:
          text = f.read()
          return text
    
#эта функция выбирает рандомное слово из списка    
def random_word(text):
    text = text.split(';')
    word = random.choice(text)
    return word

#эта функция предлагает пользователю угадать слово по буквам
def attempt(word):
    attempt_counter = 7
    letter_list = []
    hang_index = 0
    hang_list = ['|\n|\n|', '___\n|\n|\n|', '___\n| o\n|\n|', '___\n| o\n|/\n|', '___\n| o\n|/ \\\n|', '___\n| o\n|/ \\\n|/', '___\n| o\n|/ \\\n|/ \\']
    print('_ ' * len(word) + '\n')
    while attempt_counter > 0: 
     if attempt_counter == 1:
        print('У вас есть {} попытка, чтобы угадать слово из {} букв.'.format(attempt_counter, len(word)))
     elif attempt_counter >= 5:
        print('У вас есть {} попыток, чтобы угадать слово из {} букв.'.format(attempt_counter, len(word)))
     else:
         print('У вас есть {} попытки, чтобы угадать слово из {} букв.'.format(attempt_counter, len(word)))

#ввод буквы и проверка на правильность ввода
     letter = input('Введите букву: ').lower()
     while not letter.isalpha() or len(letter) > 1:
      letter = input('Введенный символ должен быть буквой: ').lower()
     while letter in letter_list:
      letter = input('\n\nЭта буква уже была названа, введите другую: ').lower()
     letter_list.append(letter)
     guess = []
     
#проверяем,есть ли введенная буква в слове
     if letter in word:
        print('Буква "{}" есть в этом слове!'.format(letter))
     else:
         print('Такой буквы нет!\n')
         print(hang_list[hang_index])
         hang_index += 1
         attempt_counter -= 1
         
#выводим угаданные буквы
     for letter in word:
        if letter in letter_list:
            guess.append(letter + ' ')
        elif letter not in letter_list:
            guess.append('_ ')
     print('\n' + ''.join(guess) + '\n\n')
     
#инструкции для победы и поражения     
     if re.sub(' ', '', ''.join(guess)) == word:
         print('Вы выиграли!')
         break
    else:
        print('Вы проиграли!\nЭто было слово: ' + word)

#предлагаем сыграть ещё раз        
def once_again():
        once_again = input('Хотите сыграть ещё раз? Напишите "да" или "нет": ').lower()
        while once_again != 'да' and once_again != 'нет':
             once_again = input('Пожалуйста, ответьте: ').lower()
        if once_again == 'нет':
            print('Спасибо, что сыграли!')
        else:
             main()
          
              
def main():
    text = choose_topic()
    word = random_word(text)
    fun = attempt(word)
    once_again()
    
if __name__ == '__main__':
    main()
```

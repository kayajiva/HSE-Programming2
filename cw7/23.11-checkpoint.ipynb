```python
import matplotlib
import matplotlib.pyplot as plt

# X и Y - координаты точек
X = [1,   1.5, 2,   3, 4, 6, 5]
Y = [1.8, 1,   3.6, 2, 42, 3, 45]


plt.scatter(X,Y, c = ['orchid', 'blue'], s = [70, 20], marker = 'H')  # рисуем точки
plt.title("Просто точки")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt

# X и Y - координаты точек
X = [1,   1.5, 2,   3]
Y = [1.8, 1,   3.6, 2]


dots = ['Первая', "Вторая", "Третья", "Четвертая", "Пятая", "Шестая"]
markers = ['o', '*', '8', 'D', '^', 'P']

for x, y, m, d in zip(X, Y, markers, dots):
    plt.scatter(x, y, marker=m, s=100)
    plt.text(x+0.005, y-0.2, d) # +0.1 - это чтобы текст не наползал на маркер, а отрисовывался чуть выше и правее 
plt.show()
```

```python
plt.plot(range(12), linestyle='--', marker='p', c='r', markevery=[4, 6])
plt.show()

plt.savefig('plot_with_colors.png')
```

```python
from matplotlib import style  # добавляем стили
style.use('ggplot')  # выбираем стиль ggplot

plt.plot(X,Y)
plt.title('График с ggplot')

plt.show()
```

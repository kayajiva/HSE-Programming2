```python
import networkx as nx

G = nx.Graph()
G.add_nodes_from([1, 2 ,3, 4, 5, 6, 7])
G.add_edges_from([(1, 2), (1, 3), (1, 4), (1, 5), (1, 6), (1, 7), (2, 3), (3, 5)])
print('узлы', G.nodes())
print([n for n in G.neighbors(1)])

import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot') 

pos=nx.spring_layout(G)

labeldict = {}
labeldict[1] = "двигаться"
labeldict[2] = "бежать"
labeldict[3] = "идти"
labeldict[4] = "плыть"
labeldict[5] = "ползти"
labeldict[6] = "прыгать"
labeldict[7] = "лететь"

nx.draw(G, pos, labels=labeldict, with_labels = True)
#nx.draw_networkx_nodes(G, pos, node_color='red', node_size=50, with_labels = True)
#nx.draw_networkx_edges(G, pos, edge_color='yellow',  with_labels = True) 
plt.axis('off') 
plt.show() 
```
